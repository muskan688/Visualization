```python
#Label
x=np.array([10,15,20,25,30,35,40])
y=np.array([25,36,42,55,62,56,49])

font1={'family':'serif', 'color':'green', 'size':20}

plt.plot(x,y)
plt.xlabel('Average Hours')
plt.ylabel('Average Score')
plt.title('Line Chart', fontdict=font1, loc= 'center')
plt.show()
```

```python
#grid
x=np.array([10,15,20,25,30,35,40])
y=np.array([25,36,42,55,62,56,49])

font1={'family':'serif', 'color':'green', 'size':20}

plt.plot(x,y)
plt.xlabel('Average Hours')
plt.ylabel('Average Score')
plt.title('Line Chart', fontdict=font1, loc= 'right')
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd

#sample data
x = np.linspace(0,10,100)
y = np.cos(x) + np.random.normal(0,0.1,100)   #Generates 100 random values from a normal (Gaussian) distribution with a mean of 0 and a standard deviation of 0.1.
df = pd.DataFrame({'x' : x, 'y' : y})

window_size=5
df['y_smooth'] = df['y'].rolling(window=window_size).mean()
plt.plot(df['x'], df['y'], label = 'original', alpha = 0.5)
plt.plot(df['x'], df['y_smooth'], label='smoothed', color='blue')
plt.legend()
plt.show()
```

```python
import pandas as pd


#sample data
x = np.linspace(-10, 10, 100) #selecting random 100 numbers btw -10 to 10
y = x**2 + np.random.normal(0, 10, 100)

df = pd.DataFrame({'x': x, 'y': y})

window_size = 5
df['y_smooth'] = df['y'].rolling(window = window_size).mean()

plt.plot(df['x'], df['y'], label = 'original', alpha = 0.5)
plt.plot(df['x'], df['y_smooth'], label = 'parabola', color = 'green')
plt.legend()
plt.show()
```
